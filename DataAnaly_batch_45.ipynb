```python
import pandas as pd

dataset = pd.read_csv('/content/50_Startups.csv')
dataset.head()
```

```python
dataset.info()
```

```python
# imputation -> model based, mean, median, mode
dataset['Marketing Spend'] = dataset['Marketing Spend'].fillna(dataset['Marketing Spend'].mean())
```

```python
dataset.info()
```

```python
dataset['State'] = dataset.State.fillna(dataset.State.mode()[0])
```

```python
dataset.info()
```

```python

from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
dataset['State_t'] = lb.fit_transform(dataset.State)
```
